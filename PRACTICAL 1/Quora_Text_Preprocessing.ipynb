```python
import json
import ssl
import urllib
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup
import requests
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url = input("Enter the question that you want to find answer for -\n")
url = "https://www.quora.com/" + url.replace(" ","-")
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.content, "html.parser")
Quora_json=dict()
```

```python
question = soup.find("title")
Quora_json["question"]=question.text.replace("- Quora"," ")
```

```python
answers = soup.find("script",{"type":"application/ld+json"})
ans_list = json.loads(answers.string)
ans_list
```

```python
for number, answer in enumerate(ans_list["mainEntity"]["suggestedAnswer"], start=1):
    print(f"Answer: {number}. | Upvote count: {answer['upvoteCount']}")
    print(answer["text"].strip())
    print("-" * 75)
```

```python
Quora_json["answers"]=[]
for i in range(0,10):
    answer_val= {"name": ans_list["mainEntity"]["suggestedAnswer"][i]["author"]["name"],
                 "text": ans_list["mainEntity"]["suggestedAnswer"][i]["text"],
                 "upvoteCount": ans_list["mainEntity"]["suggestedAnswer"][i]["upvoteCount"]}
   # answer_val_list.append(answer_val)
    Quora_json["answers"].append(answer_val)
```

```python
print(Quora_json)
```

```python
with open("Quora_data.json",'w')as outfile:
    json.dump(Quora_json,outfile,indent=4)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import *
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
with open("Quora_data.json", "r") as file:
    data=json.load(file)
ans=[]
for i in data["answers"]:
    ans.append([i['text']])
```

```python
def converttolower(text):
    return text.lower()
for i in range(10):
    ans[i]=[converttolower(ans[i][0])]
ans
```

```python
import string
import re
```

```python
def removeurl(text):
    return re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
for i in range(10):
    ans[i]=[removeurl(ans[i][0])]
ans
```

```python
def removepunctuation(text):
    return text.translate(str.maketrans("","",string.punctuation+"'"+'"'+"‚Äô"))
for i in range(10):
    ans[i]=[removepunctuation(ans[i][0])]
ans
```

```python
import nltk
```

```python
def tokenization(text):
    return nltk.word_tokenize(text)
for i in range(10):
    ans[i]=tokenization(ans[i][0])
ans
```

```python
def removestop(text):
    return [word for word in text if not word in stopwords.words()]
for i in range(10):
    ans[i]=removestop(ans[i])
ans
```

```python
porter = PorterStemmer()
def stemming(text):
    return [porter.stem(word) for word in text]
for i in range(10):
    ans[i]=stemming(ans[i])
ans
```

```python
lematizer=WordNetLemmatizer()
def lematizing(text):
    return [lematizer.lemmatize(word) for word in text] 
for i in range(10):
    ans[i]=lematizing(ans[i])
ans
```
